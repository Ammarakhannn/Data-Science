***Fix data inconsistencies:***

Changing the columns name:

4. VisITedResources----> Vistedresources
5. ParesntschoolStatisfaction -----> ParentSchoolStatisfaction

Kuwait is written as 'KuwaIT' in *PlaceofBirth* column and as 'KW' in *nationality* column that can cause in consistency so changing it to 'Kuwait' using replace to remove incosistency.

```python
data[['PlaceofBirth', 'Nationality']] = data[['PlaceofBirth', 'Nationality']].replace(['KW', 'KuwaIT'], 'Kuwait')
data
```

***Handle outliers:***

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

Checking Outlier in AnnoucementsView at threshold 2 on threshold 3,4 their are no outliers found and plotting the graph of outlier.

```python
value_column = 'AnnouncementsView'
outliers = data[(data[value_column] < data[value_column].mean() - 2 * data[value_column].std()) |
              (data[value_column] > data[value_column].mean() + 2 * data[value_column].std())]
print(outliers)
```

```python
sns.catplot(x= "GradeID", y = "AnnouncementsView", data = outliers, kind = "box",aspect = 1.5)
plt.title("Boxplot for GradeID vs AnnoucementsView")
plt.show()
```

Checking outlier in Discussion column at threshold 2 on threshold 3,4 their are no outliers found.

```python
value_column = 'Discussion'
outliers = data[(data[value_column] < data[value_column].mean() - 2 * data[value_column].std()) |
              (data[value_column] > data[value_column].mean() + 2 * data[value_column].std())]
print(outliers)
```

```python
sns.catplot(x= "Nationality", y = "Discussion", data = outliers, kind = "box",aspect = 1.5)
plt.title("Boxplot for Nationality vs Discussion")
plt.show()
```
