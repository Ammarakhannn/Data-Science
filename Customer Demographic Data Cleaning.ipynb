###HANDLING NULL VALUES###

---



####**Checking for null Values in the dataset**

```python
data.isnull().sum()
```

####***'last_name'*** Column

Filling the null values in ***last_name*** column with ***'Unknown'*** since the last_name are unique in most cases so imputing it with mode value will not be a good choice and deleting the columns can make are dataset short because their are many null values in last_name column.

```python
data['last_name'] = data['last_name'].fillna("Unkown");
data.isnull().sum()
```

#### ***'DOB'*** & ***'tenure'*** Column

Dropping the null values in ***DOB*** column because their are only 87 cells in the DOB column with null values and dropping them will not cause any issue as 87 is not a big number of rows in 4000 rows dataset
and by deleting the 87 rows in ***DOB*** column ***tenure*** column null values are also deleted.

```python
# print("Mode value for DOB col is ", data['DOB'].mode())
# data['DOB'] = data['DOB'].fillna("1978-01-30");
data.dropna(subset=['DOB'], inplace=True)
data.isnull().sum()
```

#### ***'job_title'*** Column

Filling the null values in ***job_title*** column with mode value that column i.e ***'Business Systems Development Analyst'*** since deleting the columns can make are dataset short because their are total of *497* null values in job_title column.

```python


print("Mode value for job_title col is ", data['job_title'].mode())
print((data['job_title'] == "Business Systems Development Analyst").sum() , "Occurrences");
print()

data['job_title'] = data['job_title'].fillna("Business Systems Development Analyst");
data.isnull().sum()
```

##### ***'job_industry_category'*** Column

Filling the null values in ***job_industry_category*** column with mode value that column i.e ***'Manufacturing'*** since deleting the columns can make are dataset short because their are total of *656* null values in job_industry_category column.

```python
print("Mode value for job_industry_category col is ", data['job_industry_category'].mode())
print((data['job_industry_category'] == "Manufacturing").sum() , "Occurrences");
print()

data['job_industry_category'] = data['job_industry_category'].fillna("Manufacturing");
data.isnull().sum()
```

#### ***'default'*** Column

Filling the null values in ***default*** column with mode value that column i.e ***'1.00E+02'*** since deleting the columns can make are dataset short because their are total of *215* null values in default column.

```python
print("Mode value for default col is ", data['default'].mode())
print((data['default'] == "1.00E+02").sum() , "Occurrences");
print()

data['default'] = data['default'].fillna("1.00E+02");
data.isnull().sum()
```

###Structuring the ***'gender'*** column

---



```python
gender_structure = {
    'M' : 'Male',
    'F' : 'Female',
    'U' : 'Unknown'
}
data['gender'] = data['gender'].replace(gender_structure)
data
```

###Derving ***'Age'*** Column from ***'DOB'*** column

---



Changing the format of DOB column to date format using to_datetime function in pandas getting the current year using datetime.now function from panda and subtracting the current year from DOB year to get the age and saving it in a new column name age.

```python
data['DOB']= pd.to_datetime(data['DOB'])
current_year = pd.datetime.now().year
print(current_year);
data['Age'] = current_year - data['DOB'].dt.year
data
```

####Deleting ***DOB*** column

```python
data.drop(['DOB'] , axis = 1 , inplace= True );
data
```

###OUTLIERS

---



####outliers found in tenure column at threshold 2

```python
value_column = 'tenure'
outliers = data[(data[value_column] < data[value_column].mean() - 2 * data[value_column].std()) |
              (data[value_column] > data[value_column].mean() + 2 * data[value_column].std())]

print("Outliers:\n")
outliers
```

####outliers found in age column at threshold 3

```python
value_column = 'Age'
outliers = data[(data[value_column] < data[value_column].mean() - 2 * data[value_column].std()) |
              (data[value_column] > data[value_column].mean() + 2 * data[value_column].std())]

print("Outliers:\n")
outliers
```
